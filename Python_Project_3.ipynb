```python
#creating a dataframe wherein its only arrest data from Queens
df_Queens = df[df['ARREST_BORO']=='Q']
df_Queens.head()
```

```python
#wanted to know the most frequent reasons for arrest in Queens
df_Queens.groupby(['PD_DESC'])['ARREST_KEY'].count().sort_values(ascending=False)
```

```python
df_Queens.groupby(['OFNS_DESC'])['ARREST_KEY'].count().sort_values(ascending=False)
```

```python
#similar to the city-wide graph above, wanted to see overall numbers for arrests broken down by race
df_Queens.groupby(['PERP_RACE'])['ARREST_KEY'].count().sort_values(ascending=False).plot(kind='bar')
```

```python
#similar to above, wanted to see category of offenses broken down by race, specific for Queens
sns.displot(data=df_Queens, x="PERP_RACE", col="LAW_CAT_CD", hue="PERP_RACE")
```

SOME TAKEAWAYS RE: CRIME IN QUEENS.  Just like the larger dataset, the most arrests involve detaining Black individuals. Unlike city-wide numbers, AAPI individuals are the 3rd most arrested group (where they rank 5th in city wide numbers).  In examining arrests by seriousness (e.g. felony, misdameanor, etc) the gap between arrests of Black individuals and Hispanics for misdameanor offenses appears smaller in Queens than city-wide numbers.  While these topline numbers do tell a story, they do not account for the underlying demographics of the city or borough of Queens overall. The nominal arrest numbers don't reflect per capita trends or rates.

City wide, the most common offenses were:  

*   ASSAULT 3 & RELATED OFFENSES
*   PETIT LARCENY
*   FELONY ASSAULT
*   DANGEROUS DRUGS
*   MISCELLANEOUS PENAL LAW

In Queens, the most common offenses were:
          
*   ASSAULT 3 & RELATED OFFENSES
*   FELONY ASSAULT  
*   PETIT LARCENY  
*   MISCELLANEOUS PENAL LAW
*   VEHICLE AND TRAFFIC LAWS

None of this analysis accounts for repeat offenders.

```python
#creating a dataframe wherein its only arrest data relates to canabis offenses
df_pot = df[df['PD_DESC'].str.contains('CANNABIS')]
df_pot.head()
```

```python
#similar to above, looking at number of arrests broken down by race (but only for cannabis offenses)
df_pot.groupby(['PERP_RACE'])['ARREST_KEY'].count().sort_values(ascending=False).plot(kind='bar')
```

```python
#similar to above, looking to see if the severity of the cannabis related offenses varies by race
sns.displot(data=df_pot, x="PERP_RACE", col="PD_DESC", hue="PERP_RACE")
```

```python
sns.displot(data=df_pot, x="PERP_RACE", col="ARREST_BORO", hue="PERP_RACE")
```

SOME SUMMARY THOUGHTS ON CANNABIS RELATED ARRESTS:  Similar to city-wide arrests of all kinds, most of the people arrested for cannabis related offenses are Black.  

When examining seriousness of offenses, city wide arrests of all kinds Black individuals were arrested more often. However, when examining cannabis related offenses, more white and AAPI individuals were arrested for cannabis possession 2&1.

Lastly, when examining arrests within each borough for all offenses, Black individuals were arrested most often. However, for cannabis related offenses, white Hispanics were arrested most often in Queens, Brooklyn, and Manhattan.

None of these trends account for repeat offenders.
